```python
from learn_vrc_tst import geom
from automol import smiles
from scipy.special import gamma
import numpy as np
import pint
```

```python
def phase_space_potential_coefficient_from_reference(mu: float, k0: float, T0: float=300, n: float=6) -> float:
    """Calculate the phase space potential coefficient for a given reduced mass, temperature, and parameters.

    Args:
        mu: Reduced mass of the system (in atomic mass units).
        k0: High-pressure rate coefficient at the reference temperature (in cm^3/s).
        T0: Reference temperature (in Kelvin).
        n: Temperature exponent for the rate coefficient.

    Returns:
        float: Phase space potential coefficient (in appropriate units).
    """
    k0 = pint.Quantity(k0, "cm^3/s").m_as("a0^3 Eh / hbar")
    mu = pint.Quantity(mu, "dalton").m_as("m_e")
    kbT0 = (pint.Quantity(T0, "K") * pint.Quantity("boltzmann_constant")).m_as("Eh")
    C = ((k0 / gamma(1 - 2/n)) * np.sqrt(mu / (8 * np.pi * kbT0)))** (n / 2) * 2 * kbT0 / (n - 2)
    return C.item()


# Determine reduced mass
r = smiles.geometry("C1C[CH]C=C1")
# k0 = 4e-12
# r = smiles.geometry("C1C[CH]CC1")
k0 = 2e-11
o2 = smiles.geometry("O=O")
m1 = r.total_mass
m2 = o2.total_mass
mu = (m1 * m2) / (m1 + m2)

print(f"{m1 = }")
print(f"{m2 = }")
print(f"{mu = }")
phase_space_potential_coefficient_from_reference(mu, k0=k0)
```
