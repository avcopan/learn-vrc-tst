Following along with the implementation detailed in the SI of [this paper](https://doi.org/10.1073/pnas.1616208113)...

**Step 1.** Set the length of the reaction coordinate `s` to `s_min`.

```python
from automol import smiles
from learn_vrc_tst import geom, coord, View
import numpy as np
from importlib import reload

reload(geom)
reload(coord)

s = s_min = 5  # Set to 10 A initially

# Frag 1
frag1 = smiles.geometry("[CH3]")
geom.translate(frag1, -frag1.coordinates[0], in_place=True)
piv1 = coord.closest_unit_perpendicular(frag1.coordinates, away=True)
piv1_start = frag1.coordinates[0]
piv1_end = np.add(piv1, piv1_start)

# Frag 2
frag2 = smiles.geometry("[CH3]")
geom.translate(frag2, -frag2.coordinates[0], in_place=True)
geom.translate(frag2, [s, 0.0, 0.0], in_place=True)
piv2 = coord.closest_unit_perpendicular(frag2.coordinates, away=True)
piv2_start = frag2.coordinates[0]
piv2_end = np.add(piv2, piv2_start)

view = View()
view.add_geometry(frag1)
view.add_geometry(frag2)
view.add_arrow(piv1_end, start_coord=piv1_start, color="red")
view.add_arrow(piv2_end, start_coord=piv2_start, color="red")
view.show()
```
