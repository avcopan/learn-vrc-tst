```python
import pint
```

```python
alpha = pint.Quantity(501, "cm**-1")
Z = pint.Quantity(5.65 * 10**-10, "cm**3 molecule**-1 s**-1")
Z * alpha
```

```python
P = pint.Quantity(1, "torr")
kb = pint.Quantity(1, "boltzmann_constant")
T = pint.Quantity(1000, "K")
n = P / (kb * T)
n = n.to("cm**-3")
```

```python
print("{:.3~P}".format(Z * n * alpha))
```
