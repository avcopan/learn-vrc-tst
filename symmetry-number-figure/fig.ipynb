```python
from xyzrender import load, render, orient, build_config
from xyzrender.types import VectorArrow
import numpy as np
from learn_vrc_tst import geom
from scipy.spatial.transform import Rotation
```

```python
file_in = "methane-with-dummies.xyz"
file_out = "methane-oriented.xyz"
```

```python
geo = geom.read_xyz_file(file_in)
geo = geom.rotate(geo, rot=Rotation.from_euler("yx", [-10, 20], degrees=True))
geom.write_xyz_file(geo, file_out)
```

```python
cfg = build_config(
  config_name="flat",
  orient=False,
  atom_scale=3,
  bond_width=25,
  bond_color="#000000",
  atom_stroke_width=3,
  hy=True,
  bo=False
)
```

```python
mol = load(file_out)
render(mol, output="methane.svg", config=cfg)
```
