```python
from xyzrender import load, render, orient, build_config
from xyzrender.types import VectorArrow
import numpy as np
from learn_vrc_tst import geom, View
from scipy.spatial.transform import Rotation
from pathlib import Path
```

```python
geo1 = geom.read_xyz_file("methylamine-in1.xyz")
geo2 = geom.read_xyz_file("methylamine-in2.xyz")

geo1 = geom.rotate(geo1, rot=Rotation.from_euler("zx", [60, -90], degrees=True))
geo2 = geom.translate(geo2, [-1.5, 0, 0])
geo2 = geom.rotate(geo2, rot=Rotation.from_euler("y", -30, degrees=True))
geo = geom.concat([geo1, geo2])

geo = geom.remove_atoms(geo, [1, 4, 8])

geom.write_xyz_file(geo, "methylamine-in.xyz")

view = View()
view.add_geometry(geo, label=True)
view.show()
```
