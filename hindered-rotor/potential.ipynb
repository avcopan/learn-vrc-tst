# Explanation

The following does a Fourier fit of a torsional potential, matching results from MESS.

```python
from learn_vrc_tst import View, geom, coord

geo = geom.read_xyz_file("geom.xyz")

view = View()
view.add_geometry(geo, label=True)
view.show()
```

```python
import numpy as np
import altair as alt
import polars as pl
from scipy.fft import rfft, irfft

phi = [
    0.00, 30.00, 60.00, 90.00, 120.00, 150.00,
    180.00, 210.00, 240.00, 270.00, 300.00, 330.00
]
V = [
    0.0000, 0.6618, 1.6029, 1.8244, 1.8056, 1.9676,
    1.8612, 1.5288, 1.7364, 2.2327, 1.9307, 0.7382
]

N_fit = 100
n_points = len(phi)
V_fit = irfft(rfft(V), N_fit) * N_fit / n_points

phi_ref, V_ref = np.loadtxt("fourier_fit_ref.txt").T

point_data = pl.DataFrame({"phi": phi, "V": V})
line_data = pl.DataFrame({"phi": phi_ref, "V": V_ref, "V_fit": V_fit})

point_chart = alt.Chart(point_data).mark_point().encode(x=alt.X("phi"), y=alt.Y("V"))
line_chart = alt.Chart(line_data).transform_fold(["V", "V_fit"]).mark_line().encode(x=alt.X("phi:T"), y=alt.Y("value:Q"), color="key:N")
(point_chart + line_chart).show()
```
